# Test de la Chi-cuadrado

```python
import pandas as pd
import numpy as np
df=pd.read_csv("C:/Users/jenny/Documents/GITHUB/Linkedin_courses_bbdd/base_datos_2008.csv")
```

```python
np.random.seed(0) #fijamos la semilla para erplicar el resultado
df = df[df["Origin"].isin(["HOU","ATL","IND"])]
df = df.sample(frac=1)
df = df[0:10000]
```

```python
df["BigDelay"] = df["ArrDelay"]>30
observados = pd.crosstab(index=df["BigDelay"],columns=df["Origin"], margins=True)
```

```python
observados
```

```python
from scipy.stats import chi2_contingency
```

```python
test = chi2_contingency(observados)
```

```python
test
```

```python
esperados = pd.DataFrame(test[3])
```

```python
esperados
```

```python
#valores teóricos esperados
esperados_rel= round(esperados.apply(lambda r: r/len(df)*100, axis=1),2)
```

```python
#valores teóricos observados
observados_rel= round(observados.apply(lambda r: r/len(df)*100, axis=1),2)
```

```python
esperados_rel
```

```python
observados_rel
```

```python
test[1]
```

### Resumen te  test de Hipótesis

* Si el p-valor< 0.05 hay diferencias significativas: Hay relación entre las variables
* So el p-valor>0.05, no hay diferencia significativas: No hay relación entre las variables

# Análisis de Datos Extremos

```python
import pandas as pd
import numpy as np
df=pd.read_csv("C:/Users/jenny/Documents/GITHUB/Linkedin_courses_bbdd/base_datos_2008.csv")
```

```python
x=df["ArrDelay"].dropnaa()
```

```python
Q1 = np.percentile(x,25)
Q2 = np.percentile(x,50)
Q3 = np.percentile(x,75)
rangointer = Q3 - Q1
```

```python
# Un criterio básico para calcular outliers es calcular umbrales
umbralsuperior = Q3 + 1.5 * rangointer
umbralinferior = Q1 - 1.5 * rangointer
```

```python
umbralsuperior
```

```python

umbralinferior
```

```python
np.mean(x>umbralsuperior)
```

```python
np.mean(x<umbralinferior)
```

# Joins. Trabajar con base de datos relacionales

```python
import pandas as pd
consumidores = [("A", "Móvil"),("B", "Móvil"),("A","Portátil"),("A","Tablet"),("B","Tablet"),("C","Portátil"),("D", "Smartwatch"),("E","Consola")]
con_labels = ["Consumidor", "Producto"]
con_df = pd.DataFrame.from_records(consumidores,columns=con_labels)
productores=[("a", "Móvil"), ("a","Smartwatch"),("a", "Tablet"), ("b", "Portátil"),("c","Sobremesa"), ("c","Portátil")]
prod_labels=["Productor","Producto"]
prod_df = pd.DataFrame.from_records(productores, columns = prod_labels)
```

```python
con_df
```

```python
prod_df
```

```python
pd.merge(con_df, prod_df, on="Producto", how="outer") #outer son comunes y no comunes, los no comunes se completan con NaN
```

```python
pd.merge(con_df, prod_df, on="Producto", how="inner") #inner son solo los comunes
```

```python
pd.merge(con_df, prod_df, on="Producto", how="left") #left son solo los de la izquierda
```

```python
pd.merge(con_df, prod_df, on="Producto", how="right") #right son solo los de la izquierda
```

# Paralelizar

```python
import pandas as pd
import numpy as np
from joblib import Parallel, delayed

df=pd.read_csv("C:/Users/jenny/Documents/GITHUB/Linkedin_courses_bbdd/base_datos_2008.csv", nrows=100000)
```

```python
df_sub = df[["CarrierDelay","WeatherDelay","NASDelay","SecurityDelay","LateAircraftDelay"]]
df_sub.head(10)
```

```python
def retraso_maximo(fila):
    if not np.isnan(fila).any():
        names=["CarrierDelay","WeatherDelay","NASDelay","SecurityDelay","LateAircraftDelay"]
        return names[fila.index(max(fila))]
    else:
        return "None"
```

```python
results = []
for fila in df_sub.values.tolist():
    results.append(retraso_maximo(fila))
```

```python
results
```

```python
result = Parallel(n_jobs=2, backend="multiprocessing")(
            map(delayed(retraso_maximo), df_sub.values.tolist()))
```

```python
result
```
